# #webscraping an ecommerce game console

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
webpage=requests.get('https://www.banggood.com/search/game-console.html?akmClientCountry=KE&from=nav')
sp=bs(webpage.content,'html.parser')
print(sp)
```

```python
#getting the required data
title=sp.find_all('a','title')
sellprice=sp.find_all('span','price')
originalprice=sp.find_all('span','price-old')
```

```python
#looping the data
titleloop=[titles.text for titles in title]
sellpriceloop=[sell.text for sell in sellprice]
originalpriceloop=[origin.text for origin in originalprice]
print(titleloop)

```

```python
#putting data in dictionaries
data={
    'Name_of_console':titleloop,
    'sellingprice':sellpriceloop,
    'original_price':originalpriceloop,}
```

```python
#length to know if we need to create two data frames then join them 
print(len(titleloop))
print(len(sellpriceloop))
print(len(originalpriceloop))
```

```python
#putting data in a dataframe
df=pd.DataFrame(data,columns=[
    'Name_of_console',
    'sellingprice'])
    
```

```python
#joining the dataframes
df2=pd.DataFrame(data,columns=['original_price'])
originalprice=df2['original_price']
df=df.join(originalprice)
print(df)
```
